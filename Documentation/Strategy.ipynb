# Workflow:

1) User inputs requirements to GUI
2) GUI structures requirements as JSON and passes them to Recipe Generator
3) Recipe Generator outputs a recipe to Validator
4) If recipe passes, send results to user. Otherwise loop back to Recipe Generator with feedback.

# Key metrics:


* Is the recipe structured correctly?
* Does the recipe meet the user's criteria?
